```python
import pandas as pd
```

```python
df = pd.read_csv(r"Flavors.csv")
```

```python
df
```

```python
group_by_frame = df.groupby('Base Flavor')
```

```python
df.info()
```

```python
group_by_frame.mean(numeric_only = True) # Need to add the argument Numeric_only to use this correctly
#Groups the Elements and gives the avg of the values
```

```python
df.groupby('Base Flavor').mean(numeric_only = True)
```

```python
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').min() #takes the minimum value 
```

```python
df.groupby('Base Flavor').max() # takes the max value for everything, including Str in alphabetical order
```

```python
df.groupby('Base Flavor').sum(numeric_only = True)
```

```python
df.groupby('Base Flavor').agg({'Flavor Rating':['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum']})
```

```python
df.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum']})
```
